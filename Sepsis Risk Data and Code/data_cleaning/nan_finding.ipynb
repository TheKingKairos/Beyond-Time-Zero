```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/MIMIC-ED/event_level_v2.csv")
```

```python
# count the nans
nan_counts = df.isna().sum()
positive_nan_counts = nan_counts[nan_counts > 0]
print(positive_nan_counts)
```

```python
df_cleaned = pd.read_csv("../data/MIMIC-ED/event_level_cleaned_v2.csv")
cleaned_nan_counts = df_cleaned.isna().sum()
positive_cleaned_nan_counts = cleaned_nan_counts[cleaned_nan_counts > 0]
print(positive_cleaned_nan_counts)
```
